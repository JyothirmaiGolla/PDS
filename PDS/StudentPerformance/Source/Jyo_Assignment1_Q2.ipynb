```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("/content/student.csv")
```

```python
df = pd.DataFrame(data)
```

```python
# 1. Bar Chart of Gender Distribution
plt.figure(figsize=(8, 4))
sns.countplot(data=df, x='gender')
plt.title('Gender Distribution')
plt.show()
```

```python
# 2. Bar Chart of Race/Ethnicity Distribution
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='race/ethnicity', order=df['race/ethnicity'].value_counts().index)
plt.title('Race/Ethnicity Distribution')
plt.show()
```

```python
# 3. Histogram of Math Scores
plt.figure(figsize=(8, 4))
sns.histplot(data=df, x='math score', bins=10, kde=True)
plt.title('Distribution of Math Scores')
plt.show()
```

```python
# 4. Scatter Plot of Reading vs. Writing Scores
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='reading score', y='writing score', hue='gender')
plt.title('Reading vs. Writing Scores')
plt.xlabel('Reading Score')
plt.ylabel('Writing Score')
plt.show()
```

```python
# 5. Stacked Bar Chart of Lunch and Test Preparation Course
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='lunch', hue='test preparation course')
plt.title('Lunch and Test Preparation Course Distribution')
plt.show()
```
